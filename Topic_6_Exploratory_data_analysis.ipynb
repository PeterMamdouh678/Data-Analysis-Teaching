## Data Analysis and Data Analytics

Data Analysis:
> It is hands-on data exploration and evaluation  
> It looks backwards, providing historical view of what has happened  
> It is not involved in prediction of what will happen

Data Analytics:
> It is a broader term and includes data analysis  
> It is the science behind the analysis  
> It is concerned with the entire methodology  

## Categorization of Analysis and Analytics


| Data Analysis        |  Data Analytics      |
|----------------------|----------------------|

> What is happening now (or happened so far).

2. There are many techniques available in diagnostic analytics.

There are many techniques available in Mechanistic analytics, among them **regression** is one of the most frequently used technique.

It involves the following steps:

3. Prescribe an optimal course of action in real time

# Correlation Analysis
 > ðŸ“Œ It describes strength and direction of the relationship between two variables. 
 
  1. Strength indicates how closely two variables are related to each other.
  2. Direction indicates how one variable would change its value as the value of the other variable changes.

## Pearson's *r* Correlation

The following formula is used for calculating the Pearson's *r* correlation:

$$
r = \frac{N \sum xy-\sum x \sum y}{\sqrt{\left[N\sum x^2-\left(\sum x\right)^2\right]\left[N\sum y^2-\left(\sum y\right)^2\right]}}
$$

where  
$r$ = Pearson's *r* correlation coefficient, $-1\leq r \leq 1$  
$N$ = number of values in each dataset/column  
$\sum xy$ = sum of the products of the paired scores/columns  
$\sum x$= sum of x column scores  
$\sum y$= sum of y column scores  
$\sum x^2$= sum of squared x column scores  
$\sum y^2$= sum of squared y column scores

**Directions:** All correlation coefficients between 0 and 1 represent positive correlations, while all coefficients between 0 and -1 are negative correlations. Positive relation means if one increase(decrease), then other will increase(decrease). Negative relation means if one increase(decrease), then other will decrease(increase).

**Strength:** The closer a correlation coefficient is to 1 or to -1, the stronger it is. Following picture [Figure-1] suggests a guideline to interpret the strength.

# Principal Component Analysis (PCA)
> ðŸ“Œ Project data into lower dimensions

1. It is a dimensionality reduction technique useful for data visualization and exploration.  
2. It is defined as orthogonal projection of the data onto a lower dimensional linear space (principal space) such that the variance of the projected data is maximized.

## 2-dimensional PCA

Consider the following 2d scenario:

## Intuition behind Principal Component Analysis (PCA)

- The idea is to find directions that capture maximum variations.
- The FIRST principal component, gives us the maximum variation on the principal direction (which is linear combination of original the columns).
- The second principal component, gives us the next maximum variation on another direction (which is perpendicular to the first principal direction). This direction is also a linear combination of original the columns.
- The third principal component, gives us the next maximum variation on another direction (which is perpendicular to the first and second principal direction). This direction is also a linear combination of original the columns.
- Similarly, all the principal components are ranked, based on how much variation they can capture.
- The number of principal component is equal to number of samples (or records) or number of dimensions (or columns) (whichever is smaller)

## Variance Maximization
> Book-2

Let $ \mathbf{x}_{i} $ for $ i=1,\ldots,N $ refers to $ N $ data points in $ m $ dimensions. Let $ \overline{\mathbf{x}} $ be the sample mean vector. Let $ \mathbf{u} $ be the projection vector. The variance of the projected data is given as:
$$ Var(\mathbf{u})= \frac{1}{N} \sum_{i=1}^{N} \left(\mathbf{u^Tx_{i}-u^T\overline{\mathbf{x}}}\right)^2$$

which is simplified as:  $Var(\mathbf{u})=  \mathbf{u^{T}Su} $

where $ \mathbf{S} $ is data covariance matrix, defined as:

$$ \mathbf{S}=\frac{1}{N} \sum_{i=1}^{N} (\mathbf{x_{i}-\overline{\mathbf{x}} })(\mathbf{x_{i}-\overline{\mathbf{x}} })^{T} $$

### Obtaining Principal Components (PC)

Following steps can be used for obtaining PCs.
1. Find a unit vector that maximizes $Var(\mathbf{u})$. This is the first PC.
2. Find next PC, *unit vector orthogonal to previously found PCs*, which maximizes $Var(\mathbf{u})$.
3. Repeat Step-2 until all *or required* PC are obtained.

Note: there are many efficient algorithms in the literature that obtains PCs. We will use sci-kit learn's **sklearn.decomposition** library to obtain PCs.

```python
#1. Read and display the data & data info.
import numpy as np
import pandas as pd
df = pd.read_csv('data/EDA-2.csv', delimiter = ',')
display(df.head(5))
display(df.info())
```

```python
# 2. Add the first two principal components to the database 
# and name the column names as *pca1* & *pca2*.

## Only numerical columns can be used to get principal components. 
# You can ENCODE all the non-numerical columns to numerical, if needed.
# Here we use the existing numerical columns as INPUT columns for obtaining principal components.
## NOTE: if you have an OUTPUT columns (given in the scenario, then do not use them as input columns)
# For example: here we assume column economy is an OUTPUT column 

num_columns = df.select_dtypes(exclude='object').columns
pca_input_columns=num_columns.drop(['economy'])

from sklearn.decomposition import PCA
pca = PCA(n_components=2) # get pca model

principalComponents = pca.fit_transform(df[pca_input_columns]) # get two components

## # Same as above in two steps
# pca.fit(df[pca_input_columns])
# principalComponents = pca.transform(df[pca_input_columns])

df['pca1'] = principalComponents[:,0]
df['pca2'] = principalComponents[:,1]


display(df)

### another approach to add new columns to an existing dataframe!
# df=pd.concat([df, pd.DataFrame(principalComponents)], axis=1) # add the two components
# df.rename(columns = {0:'pca1',1:'pca2'}, inplace = True) 

## See more about the .concat() method
# # help(pd.concat)
```

```python
# 3. How much variation is captured by each of the principal components.

# the variance captured by each principal component is available in .explained_variance_ratio_ attribute
# see what is inside the attribute by print(pca.explained_variance_ratio_)
# we can print the above attributes information as follows:

for ind,var in enumerate(pca.explained_variance_ratio_):
    print(f'The variance captured by pca {ind+1} is: {var*100: 0.2f}%')
```

```python
# 4. For the first principal component: 
#    i. What are the coefficients of the linear combination of the columns. 
#   ii. Which column has highest coefficient.

# The coefficients for each principal component is given in attribute .components_
# see what is inside the attribute by print(pca.components_)
# we can print the above attributes information as follows:

pca_1_coefficients=np.round(pca.components_[0],4)

zipList=list(zip(pca_input_columns,pca_1_coefficients))
ndf=pd.DataFrame.from_records(zipList,columns=['field','coeff'])

display(ndf.sort_values(by=['coeff'], ascending=False))  # remember to consider abs() for negative values...
```

```python
# 5. Construct a scatter plot using the first two principal components of the data. 
#Differentiate the points using *economy*. In a separate plot, differentiate them using *origin*.
#In a separate plot, differentiate them using *cylinders*.

import matplotlib.pyplot as plt
import seaborn as sns
# plt.figure()
# sns.relplot(x='pca1',y='pca2',
#             kind='scatter',alpha=0.75,
#             height=5, aspect=1,
#             data=df)
# plt.title('PCA vs Economy', fontsize = 20)
# plt.show()



plt.figure()
sns.relplot(x='pca1',y='pca2',
            hue='economy', palette=['r','b','g'],
            kind='scatter',alpha=0.75,
            height=5, aspect=1,
            data=df)
plt.title('PCA vs Economy', fontsize = 20)
plt.show()


plt.figure()
sns.relplot(x='pca1',y='pca2',
            hue='origin', palette=['r','b','g'],
            kind='scatter',alpha=0.75,
            height=5, aspect=1,
            data=df)
plt.title('PCA vs Origin', fontsize = 20)
plt.show()


plt.figure()
sns.relplot(x='pca1',y='pca2',
            hue='cylinders', palette=['r','b','g','m','c'], 
            kind='scatter',alpha=0.75,
            height=5, aspect=1,
            data=df)
plt.title('PCA vs Cylinders', fontsize = 20)
plt.show()
```

```python
#     5.i. Crosstable showing the total number of customers for *education & job* columns.
ctab = pd.crosstab(df['job'],df['education'])
display(ctab)

# help(pd.crosstab)
```

```python
#     5.ii. Pivot table displaying total *subscribers* for *marital, job & education* columns.

df['subscribed_num'] = df['subscribed'].apply(lambda x: 1 if x == 'yes' else 0)
pTable=df.pivot_table(values=['subscribed_num'], 
                       index=['marital', 'job'], 
                       columns=['education'], 
                       aggfunc=sum)  # the default aggFunc will be mean

display(pTable)
df.drop(['subscribed_num'], axis=1, inplace=True)  # the column was created just for the pivot table
```

```python
#     5.iii. Box-Plots for each numerical and categorical column (skip *subscribed* column), 
# and differentiate using *subscribed* column.


cat_columns = df.select_dtypes('object').columns.drop('subscribed')
num_columns = df.select_dtypes(exclude='object').columns


fig,axes = plt.subplots(len(cat_columns), len(num_columns), figsize=(15,50))
for c,nCol in enumerate(num_columns):
    for r,cCol in enumerate(cat_columns):
        axis=axes[r][c]
        sns.boxplot(x=cCol,y=nCol,hue='subscribed',data=df, ax=axis)
        axis.set_xticklabels(axis.get_xticklabels(), rotation=45)
plt.show()
```
